````python
def total_anount(*prices, discount = 0.0, **kwargs ):
  '''here 0.0 needs to added'''
  '''
  This function tokes the prices of pones ordered
  and returrs the total erder amourt.

  '''

  total = 0
  # computing the total erder amount
  for price in prices:
    total += price

  discounted_price = total - (discount*total) ##given the condition it will automatically substract
  additional_cashback = discounted_price - kwargs['cashback']

  # return the total anount

  return additional_cashback
````

````python
additionals = {'cashback': 5}## kwargs comes as a dictionary
print ( 'the total of all the phone price aftere discount is ',total_anount(239,563,864,765,734,764,discount=0.05, **additionals)) ## see here additionals came 2 times
## also remember before attitionals we have two stars in print aswell
````

````python
def total_anount(*prices,**kwargs):

  '''
  This function takes the prices of phones ordered
  and any other cost additions/subtractions,
  and returns the total amount, net spend,
  and rewards points earned for the order.
  '''
  total = 0
  # computing the total order amount
  for price in prices:
    total += price

  net_spend = total - additionals['discount']*total - additionals ['cashback']

  if total >= 10000:
    reward_points = 300
  elif total >= 5000:
    reward_points = 200
  elif total >= 2000:
    reward_points = 100
  else:
    reward_points = 0

  # return the total amount
  return total, net_spend, reward_points ##here we are returning all of the different prices after discount and rewords recived
````

````python
additionals={'discount':0.03,'cashback':5}
ta, td, tr = total_anount(239,563,864,765,734,764,**additionals) ## see here we have defined different actions in differnt key words (ta is total addition),td is discount and cashbacks, tr rewords received
print('The summery of the prices: \n', '/ntotal purchase: ', ta, 'total sepnt after all discount and cashback:',td,'total rewards recived: ', tr)
````

###Linspace
```Using np.linspace() function

. It has the following parameters:
o start: start of interval range. By default start = 0
o stop: end of interval range
```

````python
matrix2 = np.linspace(0,5) # by default 50 evenly spaced values will be generated between 0 and 5
matrix2
````

````python
vec = np.linspace(10,20,3)
print(vec)
````


**How are these values getting generated?**
```
The step size or the difference between each element will be decided by the following formula:
(stop- start) / (total elements - 1)
So, in this case: (5 - 0) / 49 = 0.10204082
The first value will be 0.10204082, the second valile will be 0.10204082 + 0.10204082, the third value will be 0.10204082 +0.10204082
+0.10204082, and so on.
```
